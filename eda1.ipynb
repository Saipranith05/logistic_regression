```python
x = df.drop(columns=["heart disease"])
y = df["heart disease"]
```

```python
x.head(3)
```

```python
#split the data into 80% and 20%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
x_train.head(3)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)
```

```python
x_train_scaled
```

```python
new_x_train = pd.DataFrame(x_train_scaled,columns=x.columns)
```

```python
new_x_train
```

```python
x_test_scaled
```

```python
new_x_test = pd.DataFrame(x_test_scaled,columns=x.columns)
```

```python
#new_x_train, new_x_test
#y_train, y_test
```

```python
#training == new_x_train, y_train
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(new_x_train,y_train)
```

```python
#prediction
y_pred = log.predict(new_x_train)
```

```python
# test the data, by using X_test - new_x_test
y_pred = log.predict(new_x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
print(confusion_matrix(y_test,y_pred)) # Logistic Regression
print("The accuracy of the Logistic regression is",accuracy_score (y_test,y_pred))
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
```

```python
import joblib
joblib.dump(log, "Logistic.pkl")
```
